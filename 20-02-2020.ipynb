```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>

struct Node{
	int Data;
	struct Node *Next;
}*S,*L,*C,*P;

void insertNode();
void removeNode();
void display();
void countNodes();
void middleNode();
void alternateNodes();

int main(){
	int x;
	S = L = C = P = NULL;
	while(1){
		system("cls");//clrscr()
		printf("\n Linked List Operations: ");
		printf("\n 1. Insert ");
		printf("\n 2. Remove ");
		printf("\n 3. Display ");
		printf("\n 4. Count Nodes ");
		printf("\n 5. Middle Node ");
		printf("\n 6. Alternate Nodes");
		printf("\n 7. Exit ");
		printf("\n Enter the Choice : ");
		scanf("%d",&x);
		switch(x){
			case 1: insertNode();
			break;
			case 2: removeNode();
			break;
			case 3: display();
			break;
			case 4 : countNodes();
			break;
			case 5 : middleNode();
			break;
			case 6 : alternateNodes();
			break;
			case 7: printf("\n Program is Terminating...");
			return 0;
			break;
		}
	}
	return 0;
}
void insertNode(){
	int pos,i=1;
	struct Node *nn;
	nn = (struct Node *)malloc(sizeof(struct Node));
	printf("\n Data : ");
	scanf("%d",&nn->Data);
	nn->Next = NULL;
	if(S==NULL){
		S = L = nn;
	}
	else{
		printf("\n Enter the Position : ");
		scanf("%d",&pos);//1 2 3 4 ....
		if(pos==1){
			nn->Next = S;
			S = nn;
		}
		else{
			for(C=S;i<pos;P=C,C=C->Next,i++);
			nn->Next = C;
			P->Next = nn;
		}		
	}
	printf("\n Node is Inserted...");
	getch();
}
void removeNode(){
	// Accept the Input -- Delete -- Node is not present
	int n;
	if(S==NULL){
		printf("\n Single Linked List Empty...");
	}
	else{
		printf("\n Enter the Node Value : ");
		scanf("%d",&n);
		if(n==S->Data && S->Next==NULL){
			S = L = NULL;
		}
		else if(n==S->Data){
			C = S;
			S = S->Next;
			C->Next = NULL;
			C = NULL;
		}
		else if(n==L->Data){
			for(C=S;C!=L;P=C,C=C->Next);
			L = P;
			P->Next = NULL;
			P = NULL;
			C = NULL;//free(P) free(C)
		}
		else{
			for(C=S;C!=NULL && C->Data!=n;P=C,C=C->Next);
			if(C==NULL){
				printf("\n Node is not Found");
			}
			else{
				P->Next = C->Next;
				C->Next = NULL;
				C = NULL;
			}
		}
		printf("\n Node is deleted...");
	}
	getch();
}
void display(){
	if(S==NULL){
		printf("\n Linked List is Empty");
	}
	else{
		printf("\n Linked List : ");
		printf("\n -------------- \n");
		for(C=S;C!=NULL;C=C->Next)
		{
			printf("%d ",C->Data);
			//printf("%u ",C->Next);
		}
	}
	getch();
}
void countNodes(){
	int cnt=0;
	if(S==NULL){
		printf("0");
	}
	else{
		for(C=S;C!=NULL;C=C->Next){
			cnt++;
		}
		printf("%d ",cnt);
	}
	getch();
}

void middleNode(){
	int cnt=0,x,i=1;
	for(C=S;C!=NULL;C=C->Next){
		cnt++;
	}
	if(cnt%2==0){
		x = cnt/2;
	}
	else{
		x = (cnt+1)/2;
	}
	for(C=S;i<x;C=C->Next,i++);
	printf("%d ",C->Data);
	getch();
}
void alternateNodes(){
	int aa=0;
	for(C=S;C!=NULL;C=C->Next,aa++){
		if(aa%2==0){
			printf("%d",C->Data);
		}
		else{
			printf("->");	
		}
	}
	printf("%d",C->Data);
	getch();
}
```

```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>
struct Node{
	int Data;
	struct Node *Next;
}*S,*L,*P,*C;

void pushNode();
void popNode();
void displayNodes();

int main(){
	int x;
	S = L = P = C = NULL;
	while(1){
		printf("\n Stack Operations Using LL : ");
		printf("\n 1. Push ");
		printf("\n 2. Pop ");
		printf("\n 3. Display ");
		printf("\n 4. Exit ");
		printf("\n Enter Your Choice : ");
		scanf("%d",&x);
		switch(x){
			case 1 : pushNode();
			break;
			case 2 : popNode();
			break;
			case 3 : displayNodes();
			break;
			case 4: printf("\n Program is Terminating...");
			return 0;
			break;
		}
	}
	return 0;
}
void pushNode(){
	struct Node *nn;
	nn = (struct Node *)malloc(sizeof(struct Node));
	printf("\n Enter the Data : ");
	scanf("%d",&nn->Data);
	nn->Next = NULL;
	if(S==NULL){
		S = L = nn;
	}
	else{
		nn->Next = S;
		S = nn;
	}
	printf("\n Node is Pushed to Stack ...");
	getch();
}
void popNode(){
	if(S==NULL){
		printf("\n Stack is Empty \n Stack is Underflow");
	}
	else{
		C = S;
		S = S->Next;
		C->Next = NULL;
		C = NULL;
	}
	printf("\n Node is Pop from Stack ...");
	getch();
}
void displayNodes(){
	if(S==NULL){
		printf("\n Stack is Empty \n Stack is Underflow");
	}
	else{
		printf("\n Stack Values are : ");
		printf("\n ------------------ \n");
		for(C=S;C!=NULL;C=C->Next){
			printf("%d ",C->Data);
		}
	}
	getch();
}

```
